```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
import matplotlib.figure as fig
```

```python
data_f1_h2 = pd.read_csv('Files/Output_0_2.csv')
data_f2_h2 = pd.read_csv('Files/Output_1_2.csv')
data_f3_h2 = pd.read_csv('Files/Output_2_2.csv')

data_f1_h3 = pd.read_csv('Files/Output_0_3.csv')
data_f2_h3 = pd.read_csv('Files/Output_1_3.csv')
data_f3_h3 = pd.read_csv('Files/Output_2_3.csv')

algorithms = ["BFS", "DFS", "Greedy", "A*"]
```

```python
def graph(x_array, data):
    times = data['time']
    v_nodes = data['v_nodes']
    f_nodes = data['f_nodes']
    moves = data['path_len']
    parameters = {'xtick.labelsize': 12,'ytick.labelsize': 12, 'axes.labelsize': 14}
    plt.rcParams.update(parameters)
    plt.figure(figsize=(7,5))
    plt.bar(x_array, times, color="#81c7a5")
    plt.ylabel("Tiempo [s]")
    plt.show()
    parameters = {'xtick.labelsize': 12,'ytick.labelsize': 12, 'axes.labelsize': 14}
    plt.rcParams.update(parameters)
    plt.figure(figsize=(7,5))
    plt.bar(x_array, moves, color="#81c7a5")
    plt.ylabel("Movimientos")
    plt.show()
    parameters = {'xtick.labelsize': 12,'ytick.labelsize': 12, 'axes.labelsize': 14}
    plt.rcParams.update(parameters)
    plt.figure(figsize=(7,5))
    plt.bar(x_array, f_nodes, color="#81c7a5")
    plt.ylabel("Cantidad de nodos frontera")
    plt.show()
    parameters = {'xtick.labelsize': 12,'ytick.labelsize': 12, 'axes.labelsize': 14}
    plt.rcParams.update(parameters)
    plt.figure(figsize=(7,5))
    plt.bar(x_array, v_nodes, color="#81c7a5")
    plt.ylabel("Cantidad de nodos visitados")
    plt.show()
```

```python
graph(algorithms, data_f1_h2)
```

```python
graph(algorithms, data_f1_h3)
```

```python
graph(algorithms, data_f2_h2)
```
